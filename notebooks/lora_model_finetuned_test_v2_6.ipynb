```python
import torch
from peft import PeftModel
from transformers import AutoModel
model_type=  "openbmb/MiniCPM-V-2_6"   # or openbmb/MiniCPM-Llama3-V-2_5 , openbmb/MiniCPM-V-2
path_to_adapter="/mnt/resource/blobfuse2tmp/nando-mlops-vision-prod/models/office/nando-eye/experiments/minicpm-v-2_6/20240903_1/output__lora/"

torch.manual_seed(0)

model = AutoModel.from_pretrained(
        model_type,
        trust_remote_code=True,
        torch_dtype=torch.float16
        )
```

```python
lora_model = PeftModel.from_pretrained(
    model,
    path_to_adapter,
    device_map="auto",
    trust_remote_code=True
).eval().cuda()
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('openbmb/MiniCPM-V-2_6', trust_remote_code=True)
```

```python
import torch
from PIL import Image
from transformers import AutoModel, AutoTokenizer

image = Image.open('21517f0363581d7d_02_09_2024__19_01_05.jpg').convert('RGB')

# First round chat 
question = "Which object is shown to the camera? Answer the question using a single word or phrase."
msgs = [{'role': 'user', 'content': [image, question]}]

answer = model.chat(
    image=None,
    msgs=msgs,
    tokenizer=tokenizer
)
print(answer)
```
